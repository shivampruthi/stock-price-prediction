```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load the data
# For this example, let's generate some random stock prices

np.random.seed(0)
days = np.arange(1, 101)
stock_prices = 100 + 2 * days + np.random.normal(0, 10, 100)

# Create a DataFrame
data = pd.DataFrame({'Days': days, 'Stock Prices': stock_prices})

# Split the data into training and testing sets
train_data = data.head(80)  # Use the first 80 days for training
test_data = data.tail(20)   # Use the last 20 days for testing

# Prepare the features and target variables
X_train = train_data[['Days']]
y_train = train_data['Stock Prices']
X_test = test_data[['Days']]
y_test = test_data['Stock Prices']

# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict stock prices
predicted_prices = model.predict(X_test)

# Plot the actual and predicted stock prices
plt.figure(figsize=(12,6))
plt.plot(days, stock_prices, label='Actual Stock Prices')
plt.plot(test_data['Days'], predicted_prices, label='Predicted Stock Prices')
plt.xlabel('Days')
plt.ylabel('Stock Prices')
plt.legend()
plt.title('Stock Price Prediction using Linear Regression')
plt.show()
```
